```python
for index, row in data.iterrows():  
    stock_name = index[1]
    plt.plot(features, row[features].values, label=stock_name)  
  
# 添加图例  
plt.legend()  
  
# 设置标题和轴标签  
plt.title('Stock Features Comparison')  
plt.xlabel('Feature Name')  
plt.ylabel('Value')  
  
# 旋转x轴标签，以便它们不会重叠  
plt.xticks(rotation=90)  
  
# 显示图表  
plt.tight_layout()  
plt.show()
```
